````python
import folium

m = folium.Map(location=[41.1537, -81.3579], zoom_start=5)

folium.Circle(
    location=[41.4993, -81.6944],
    radius=1000,
    color="blue",
    fill=True,
    fill_color="blue"
).add_to(m)

folium.Marker(
    location=[41.1537, -81.3579],
    popup="Kent",
    icon=folium.Icon(icon="bell", color="blue")
).add_to(m)

folium.Marker(
    location=[41.0814, -81.5190],
    icon=folium.DivIcon(
        html="""<div style='color:red;font-weight:bold;'>Akron!</div>"""
    )
)

m.save("simple-kent-map.html")
m
````
